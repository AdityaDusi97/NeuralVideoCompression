```python
import numpy as np
import cv2
import pdb
import base64
import matplotlib.pyplot as plt
```

```python
camera = cv2.VideoCapture('/Users/aditya/Downloads/Video_data/IMG_6269.MOV')
```

```python
while True:
    try:
        ret, frame = camera.read()
        frame = cv2.resize(frame, (640, 480))
        pdb.set_trace()
        encoded, buf = cv2.imencode('.jpg', frame)
        image = base64.b64encode(buf)
        raw_image = base64.b64decode(image)
        image1 = np.frombuffer(raw_image, dtype=np.uint8)
        frame1 = cv2.imdecode(image1, 1)
        print((frame-frame1).sum())
    except KeyboardInterrupt:
        break
```

```python
plt.imshow(abs(frame - frame1))
plt.colorbar()
```
